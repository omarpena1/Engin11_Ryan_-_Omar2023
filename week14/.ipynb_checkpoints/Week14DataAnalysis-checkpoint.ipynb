```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv("2023-04-21OutdoorData.csv")
```

a. Time/Position dependence of radiation, temperature, and 2.5 PM concentration

```python
data.plot.scatter("Latitude", "Radiation (cps)")
plt.show()
```

```python
data.plot.scatter("Latitude", "Temperature (C)")
plt.show()
```

```python
data.plot.scatter("Latitude", "Air Quality PM 2.5 (ug/m3)")
plt.show()
```

```python
data.plot.scatter("Epoch time", "Radiation (cps)")
plt.show()
```

```python
data.plot.scatter("Epoch time", "Temperature (C)")
plt.show()
```

```python
data.plot.scatter("Epoch time", "Air Quality PM 2.5 (ug/m3)")
plt.show()
```

b. Histograms

```python
data.hist("Radiation (cps)", grid = False)
plt.show()
```

```python
print("Radiation Average:", np.mean(data["Radiation (cps)"]))
print("Radiation Standard Deviation:", np.std(data["Radiation (cps)"]))
```

The radiation data does seem like it follows a normal distribution, as is expected in counting statistics

```python
data.hist("Temperature (C)", grid = False)
plt.show()
```

```python
print("Temperature Average:", np.mean(data["Temperature (C)"]))
print("Temperature Standard Deviation:", np.std(data["Temperature (C)"]))
```

The temperature data looks approximately normal, but has some peaks at the tails for some reason. This discrepancy is probably because the data was not measured at a constant temperature. 

```python
data.hist("Air Quality PM 2.5 (ug/m3)", grid = False)
plt.show()
```

```python
print("AQ Average:", np.mean(data["Air Quality PM 2.5 (ug/m3)"]))
print("AQ Standard Deviation:", np.std(data["Air Quality PM 2.5 (ug/m3)"]))
```

The air quality does not seem to follow a normal distribution at all. This is probably due to the varying levels of air quality, how the data only exists in discreet values, and perhaps just general inaccuracy.

c. Uncertainty

```python
print("Radiation Expected Variation:", np.sqrt(np.mean(data["Radiation (cps)"])))
print("Radiation Standard Deviation:", np.std(data["Radiation (cps)"]))
print("Temperature Expected Variation:", np.sqrt(np.mean(data["Temperature (C)"])))
print("Temperature Standard Deviation:", np.std(data["Temperature (C)"]))
print("AQ Expected Variation:", np.sqrt(np.mean(data["Air Quality PM 2.5 (ug/m3)"])))
print("AQ Standard Deviation:", np.std(data["Air Quality PM 2.5 (ug/m3)"]))
```

Radiation and air quality have standard deviations that are actually quite close to our expected variaiton from counting statistics, but temperature has a much lower standard deviation.

d. CLT uncertainty

```python
print("SD uncertainty:", np.sqrt(len(data.index))/(np.sqrt(2 * np.sqrt(len(data.index)))))
```

e. uncertainty analysis

For both the radiation data and the air quality data, we see that the measured standard deviation is within the uncertainty of the SD by the CLT, but the temperature data is a bit off. For the radiation data, this implies that the sources stayed very consistent throughout the measurement.
