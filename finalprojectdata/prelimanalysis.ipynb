```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
datalist = [("Etch_Soda_JacobsBackground.csv", "Etch_basement.csv"), ("Etch_Soda_JacobsBackground.csv", "Jacobs_basement.csv"), 
           ("Etch_Soda_JacobsBackground.csv", "Soda_basement.csv"), ("EvansBackground.csv", "Evans_basement.csv"), 
           ("GladeBackground.csv", "Moffitt_basement.csv"), ("GladeBackground.csv", "Mainstacks.csv"), ("SocialsciencesBackground.csv", "Socialsciences_basement.csv"),
           ("underhillbackground.csv", "underhillbasement.csv"), ("unit2background.csv", "unit2basement.csv")]
```

```python
for tup in datalist:
    print("Radiation",tup[1], tup[0], np.mean(pd.read_csv(tup[1])["Radiation (cps)"]) - np.mean(pd.read_csv(tup[0])["Radiation (cps)"]))
    print("Bismuth",tup[1], tup[0], np.mean(pd.read_csv(tup[1])["Radiation Bi (cps)"]) - np.mean(pd.read_csv(tup[0])["Radiation Bi (cps)"]))
    a = pd.read_csv(tup[0])
    b = pd.read_csv(tup[1])
    plt.figure()
    plt.hist(a["Radiation (cps)"], alpha = .75, label = "Outside")
    plt.hist(b["Radiation (cps)"], alpha = .75, label = "Inside")
    plt.xlabel("counts per second")
    plt.ylabel("frequency")
    plt.title("Radiation (cps)")
    plt.legend(loc="best")
    plt.show()
    plt.figure()
    plt.hist(a["Radiation Bi (cps)"], alpha = .75, label = "Outside")
    plt.hist(b["Radiation Bi (cps)"], alpha = .75, label = "Inside")
    plt.xlabel("counts per second")
    plt.ylabel("frequency")
    plt.title("Bismuth Radiation (cps)")
    plt.legend(loc="best")
    plt.show()
```
