# Plotting Indoor Data

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm
```

```python
myIn = pd.read_csv("feb172023insidedata.csv")
myIn = myIn.drop("Unnamed: 9", axis = 1)
myIn.head()
```

```python
data = myIn["Concentration Units (Standard) PM 1.0"]
plt.hist(data, bins = 2, density = True)
plt.title("Concentration Units (Standard) PM 1.0")
plt.xlabel("Concentration Units")
plt.ylabel("Frequency (decimal)")
mu = np.mean(data)
sigma = np.std(data)
print("Mean", mu)
print("Standard Deviation", sigma)
x = np.linspace(mu - 5*sigma, mu + 5*sigma, 100)
y = norm.pdf(x, mu, sigma)
plt.plot(x,y)
plt.show()
```

It's really cool that the data ended up being a perfect split of 1s and 0s. However since the data basically just becomes a single chunk, the distribution is more uniform than normal. Also note that it is impossible for concentration units to be negative, so a normal distribution could never be 100% accurate.

```python
data = myIn["Concentration Units (Standard) PM 2.5"]
plt.hist(data, bins = 3, density = True)
plt.title("Concentration Units (Standard) PM 2.5")
plt.xlabel("Concentration Units")
plt.ylabel("Frequency (decimal)")
mu = np.mean(data)
sigma = np.std(data)
print("Mean", mu)
print("Standard Deviation", sigma)
x = np.linspace(mu - 5*sigma, mu + 5*sigma, 100)
y = norm.pdf(x, mu, sigma)
plt.plot(x,y)
plt.show()
```

Here we can see an underlying normal distribution is a bit more apparent.

```python
data = myIn["Concentration Units (Standard) PM 10.0"]
plt.hist(data, bins = 3, density = True)
plt.title("Concentration Units (Standard) PM 10.0")
plt.xlabel("Concentration Units")
plt.ylabel("Frequency (decimal)")
mu = np.mean(data)
sigma = np.std(data)
print("Mean", mu)
print("Standard Deviation", sigma)
x = np.linspace(mu - 5*sigma, mu + 5*sigma, 100)
y = norm.pdf(x, mu, sigma)
plt.plot(x,y)
plt.show()
```

Very similar to the previous graph, but it's much more apparent that the normal distribution isn't a great model because concentration units cannot be negative.

```python
data = myIn["temperature"]
plt.hist(data, bins = 25, density = True)
plt.title("Temperature")
plt.xlabel("Temperature (Celsius)")
plt.ylabel("Frequency (%)")
mu = np.mean(data)
sigma = np.std(data)
print("Mean", mu)
print("Standard Deviation", sigma)
x = np.linspace(mu - 5*sigma, mu + 5*sigma, 100)
y = norm.pdf(x, mu, sigma)
plt.plot(x,y)
plt.show()
```

This data is not well suited for a normal distribution. There seems to be a hard cap on temperature readings and temperature also stays relatively constant.

```python
data = myIn["gas"]
plt.hist(data, bins = 20, density = True)
plt.title("Gas")
plt.xlabel("Gas")
plt.ylabel("Frequency (%)")
mu = np.mean(data)
sigma = np.std(data)
print("Mean", mu)
print("Standard Deviation", sigma)
x = np.linspace(mu - 5*sigma, mu + 5*sigma, 100)
y = norm.pdf(x, mu, sigma)
plt.plot(x,y)
plt.show()
```

This data seems a little better suited to a normal distribution than time, but for some reason there also seems to be a hard cap on the amount of gas measured. 

```python
data = myIn["pressure"]
plt.hist(data, bins = 15, density = True)
plt.title("Pressure")
plt.xlabel("Pressure")
plt.ylabel("Frequency (%)")
mu = np.mean(data)
sigma = np.std(data)
print("Mean", mu)
print("Standard Deviation", sigma)
x = np.linspace(mu - 5*sigma, mu + 5*sigma, 100)
y = norm.pdf(x, mu, sigma)
plt.plot(x,y)
plt.show()
```

The normal distribution models this data very well. This is probably because there are only very minute variations in pressure being measured.

```python
data = myIn["humidity"]
plt.hist(data, bins = 10, density = True)
plt.title("Humidity")
plt.xlabel("Humidity")
plt.ylabel("Frequency (%)")
mu = np.mean(data)
sigma = np.std(data)
print("Mean", mu)
print("Standard Deviation", sigma)
x = np.linspace(mu - 5*sigma, mu + 5*sigma, 100)
y = norm.pdf(x, mu, sigma)
plt.plot(x,y)
plt.show()
```

The normal distribution does not fit this data very well. It seems there are some anomalously high readings, but a lot of the data is concentrated in the center.

```python
data = myIn["altitude"]
plt.hist(data, bins = 15, density = True)
plt.title("Altitude")
plt.xlabel("Altitude")
plt.ylabel("Frequency (%)")
mu = np.mean(data)
sigma = np.std(data)
print("Mean", mu)
print("Standard Deviation", sigma)
x = np.linspace(mu - 5*sigma, mu + 5*sigma, 100)
y = norm.pdf(x, mu, sigma)
plt.plot(x,y)
plt.show()
```

The normal distribution fits this data well, probably for similar reasons to the pressure data. We were not moving around the sensor and it was just sitting on the table so there are only minute variations in altitude.

```python
data1 = myIn["Concentration Units (Standard) PM 1.0"]
data2 = myOut["Concentration Units (Standard) PM 1.0"]
plt.hist(data1, bins = 2, density = True, label = "Inside", alpha = .75)
plt.hist(data2, bins = 2, density = True, label = "Outside", alpha = .75)
plt.title("Concentration Units (Standard) PM 1.0")
plt.xlabel("Concentration Units")
plt.ylabel("Frequency (decimal)")
mu1 = np.mean(data1)
sigma1 = np.std(data1)
mu2 = np.mean(data2)
sigma2 = np.std(data2)
meanun = sigma1/np.sqrt(300)
x1 = np.linspace(mu1 - 5*sigma1, mu1 + 5*sigma1, 100)
y1 = norm.pdf(x1, mu1, sigma1)
plt.plot(x1,y1, label = "Inside Gaussian")
x2 = np.linspace(mu2 - 5*sigma2, mu2 + 5*sigma2, 100)
y2 = norm.pdf(x2, mu2, sigma2)
plt.plot(x2,y2, label = "Outside Gaussian")
plt.legend(loc = "best")
print("Mean Inside", mu1)
print("Standard Deviation Inside", sigma1)
print("Mean Outside", mu2)
print("Standard Deviation Outside", sigma2)
print("Mean Uncertainty", meanun)
print("Inside and Outside are seperated by", abs((mu2 - mu1)/meanun) ,"Inside Mean Standard Deviations")
plt.show()
```

The means and standard deviations are quite close, but both distributions don't seem to be well described by a Gaussian. This is probably because concentration units are discrete and not continuous. Also the difference in the distributions is probably due to the inside and outside environment having different particles.

# Comparing Sensors

```python
theirIn = pd.read_csv("otherinside.csv")
theirIn = theirIn.drop("Unnamed: 0", axis = 1)
theirIn.head()
```

```python
theirOut = pd.read_csv("otheroutside.csv")
theirOut = theirOut.drop("Unnamed: 0", axis = 1)
theirOut.head()
```

```python
data1 = myIn["Concentration Units (Standard) PM 1.0"]
data2 = theirIn["PM 1.0"]
plt.hist(data1, bins = 2, density = True, label = "Ours Inside", alpha = .75)
plt.hist(data2, bins = 2, density = True, label = "Others Inside", alpha = .75)
plt.title("Concentration Units (Standard) PM 1.0")
plt.xlabel("Concentration Units")
plt.ylabel("Frequency (decimal)")
mu1 = np.mean(data1)
sigma1 = np.std(data1)
mu2 = np.mean(data2)
sigma2 = np.std(data2)
meanun = sigma1/np.sqrt(300)
x1 = np.linspace(mu1 - 5*sigma1, mu1 + 5*sigma1, 100)
y1 = norm.pdf(x1, mu1, sigma1)
plt.plot(x1,y1, label = "Our Gaussian")
x2 = np.linspace(mu2 - 5*sigma2, mu2 + 5*sigma2, 100)
y2 = norm.pdf(x2, mu2, sigma2)
plt.plot(x2,y2, label = "Other Gaussian")
plt.legend(loc = "best")
print("Our Mean", mu1)
print("Our Standard Deviation", sigma1)
print("Other Mean", mu2)
print("Other Standard Deviation", sigma2)
print("Mean Uncertainty", meanun)
print("Our and Other are seperated by", abs((mu2 - mu1)/meanun) ,"Inside Mean Standard Deviations")
plt.show()
```

The other group's data seems to only be reading zeroes, which leads me to believe that our group's sensor had better accuracy but worse precision.

# Correlations in Data

```python
myOut.plot.scatter("temperature", "humidity")
plt.title("Temperature vs. Pressure")
plt.show()
```

The hidden variable important when making these plots is time, because we want to ensure that we're matching a temperature value to a corresponding humidity value.

It does seem the temperature is negatively correlated with humidity. We can quantify this using the correlation coefficient:

```python
print("Correlation Coefficient between temperature and humidity", myOut.corr()["temperature"]["humidity"])
```

This indicates a strong negative correlation between temperature and humidity.
